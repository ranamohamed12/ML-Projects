```python
# filling null values in training set 
def bar_chart(feature):
    survived = train_set[train_set['Survived']==1][feature].value_counts()
    dead = train_set[train_set['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('Embarked')
```

```python
# filling the missing data in fare column 
titanic['Fare'].fillna(titanic.groupby('Pclass')['Fare'].transform('median'),inplace=True)
titanic.head(10)
```
