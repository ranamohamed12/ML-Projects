**Splitting the dataset into features and target :**

```python
X = df.drop(columns=['median_house_value'])
y = df['median_house_value']
```

**Handeling incosistencies :**

```python
# Handeling numerical data :
from sklearn.impute import SimpleImputer

num_atts = [x for x in list(X.columns) if X[x].dtype != 'object']
X_num = X[num_atts]
imp = SimpleImputer(strategy='median')
imp.fit(X_num)
X_num_imputed = imp.transform(X_num) # The result is an ndarray
```

```python
X_num_imputed = pd.DataFrame(X_num_imputed, columns=num_atts, index=X.index)
```

```python
# Handeling categorical data :
X_cat = X[['ocean_proximity']]
# Encoding data into a numerical form :
from sklearn.preprocessing import OrdinalEncoder
o_encoder = OrdinalEncoder()
X_cat_encoded = o_encoder.fit_transform(X_cat)
```

```python
X_cat = pd.DataFrame(X_cat_encoded , columns=['Ocean_proximity'])
X_cat
```

But the problem with Ordianl Encoder is that the ML model will assume that close values are more related than distant ones, the solution is to use a one hot encoder

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder()
X_cat_1hot = one_hot.fit_transform(X_cat) # The result is a scipy sparse matrix
```

```python
X_cat_1hot.toarray()
```

**Using piplines :**

```python
# A pipeline is a combination of multiple preprocessing steps aligned sequentially :
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipline_steps = [
    ('inputer' , SimpleImputer(strategy='median')),
    ('scaler ' , StandardScaler())
]

num_pipline = Pipeline(num_pipline_steps)
X_num_prep =num_pipline.fit_transform(X_num)
```

```python
X_num_prep = pd.DataFrame(X_num_prep , columns=X_num.columns , index=X_num.index)
X_num_prep
```

```python
# We can combine different pipelines :
from sklearn.compose import ColumnTransformer
cat_atts = ['ocean_proximity']
transformations = [
    ("num" , num_pipline , num_atts),
    ("cat" , OneHotEncoder() , cat_atts)
]

full_pipline = ColumnTransformer(transformations)
X_prep = full_pipline.fit_transform(X) # The result is either a sparse or a dense matrix, depending on the combination
```

**Further splitting data :**

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X_prep , y , test_size=0.2 , random_state=42)
```

**Training a linear model :**

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train , y_train)
```

```python
from sklearn.metrics import mean_squared_error
predictions = lin_reg.predict(X_test)
print(f'RMSE = {np.sqrt(mean_squared_error(y_test , predictions))}')
```

**Better evaluation using cross-validation :**

Cross-validation is a statistical method used in machine learning to evaluate the performance of a model. It involves dividing the dataset into multiple parts, or folds, and then systematically using different folds for training and testing the model. This process helps to ensure that the model is not just memorizing the data (a problem known as overfitting), but is actually learning patterns that can generalize to new, unseen data12.



```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lin_reg,X_train ,y_train , scoring ='neg_mean_squared_error',cv=10)
rmse = np.sqrt(-scores)
```

```python
rmse_rounded=[]
for s in rmse:
    rmse_rounded.append(round(s,4))
print(rmse_rounded)
print(f"std = {round(np.std(rmse_rounded),4)}")
print(f"mean = {round(np.mean(rmse_rounded),4)}")
```

```python
plt.plot(range(len(rmse_rounded)) , rmse_rounded , 'r.-' , label ='training scores' , ms = 10)
plt.xlabel('Folds')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

**Training a decision tree  :**

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train , y_train)
predictions2 = tree_reg.predict(X_test)
print(f'RMSE = {round(np.sqrt(mean_squared_error(y_test , predictions2)),4)}')
```

## Training Random Forest :

```python
from sklearn.ensemble import RandomForestRegressor
rand_forest_reg  = RandomForestRegressor()
rand_forest_reg.fit(X_train , y_train)
predictions3 = rand_forest_reg.predict(X_test)
print(f'RMSE = {round(np.sqrt(mean_squared_error(y_test , predictions3)),4)}')
```
